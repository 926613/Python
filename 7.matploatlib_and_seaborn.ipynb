# Matplotlib :
#####  -is a library in pyhton for data visualisation and allows users to create variuos plots and charts to analyse data efficently.
##### - base library
##### - low level plotting library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Figure Class
##### heplful for complex visualisation with differents charts in one display

```python
fig, ax = plt.subplots()
ax.set_title('Basis figures and axis')
ax.set_xlabel('Days')
ax.set_ylabel('NUmbers of patients')
plt.show()
```

## Axes Class
##### represents individuals graphs insise figure

## Line 
##### display data points connecting by lines

```python
month=  ['Jan', 'Feb', 'Mar', 'Ap', 'May', 'June']
recovery_rate = [70, 60, 55, 80, 70, 90]

plt.plot(month, recovery_rate, marker= 'o', linestyle = '-', color = 'g', label= 'Recovery_rate')
plt.xlabel('Months')
plt.ylabel('Percentage')
plt.title('Patient Recovery Rate')
plt.legend()
plt.show()
```

```python
# observations 
# trends of recovery patient is going downwards from jan to mar and then from mar to june it is going to upward except may it is going to downwards.
```

## Bar Plot
##### it represents a categorical data in rectangular bars

```python
medicines  = ['Paracetamol', 'Antibiotics', 'Painkiller', 'Insulin']
demand = [200, 500, 160, 340]

plt.bar(medicines, demand, color = ['red', 'purple', 'blue', 'yellow'])

plt.title('Medicine Demand')
plt.xlabel('Medicine')
plt.ylabel('Units Sold')
plt.show()
```

```python
# observations : 
# here, antibiotics are more in demand than other any (unit sold around 700)
# painkiller is least in demand (around 160)
# rest of medicine are sold around(200 - 340)
```

## Pie Chart
##### shows a proportion of differnet category within a whole
##### used for percentage distribution

```python
diseases = ['Diabetes', 'Cancer', 'Asthma', 'Heart diseases', 'Others']
percentages = [20, 40, 17, 10, 13]
colors = ['yellow', 'blue', 'green', 'red', 'grey']
plt.pie(percentages, labels = diseases, autopct = '%.1f%%', colors = colors)
plt.title('Helthcare Analysis in city')
plt.show()
```

```python
# observations
# After lookin pie chart, we conclude that cancer has the maximum number of patient (around 40%)
# least diseases with percentage is heart diseases
# 60% people's face cancer and diabetes
```

## Scatter plot
##### identify the relationships between two variables
##### used for correlation analysis(advertising spend vs revenue)

```python
time_spend = [10, 15, 20, 25, 30, 35]
diagnosis_accuracy = [60, 65, 70, 75, 80, 85]

plt.scatter(time_spend, diagnosis_accuracy, color = 'red')
plt.xlabel('Time spent by doctor')
plt.ylabel('Diagnosis Accuracy')
plt.title('Time VS Diagnosis Accuracy')
plt.show()
```

```python
# here, trend is upward because accuracy is increasing with time
```

## Relational Plot
##### shows relationship between two or more variables
##### includes scater and line plot

```python
df = pd.read_csv('telecom_customer_data.csv')
df
```

```python
df.shape
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(data = df, x = 'Call_Duration', y = 'Satisfaction_Score', hue = 'Network_Provider')
# hue = 'Network_Provider' 
# Points get different colors for each network provider
# Helps compare providers easily.

plt.legend(title = 'Network provider', bbox_to_anchor = (1, 1), loc = 'upper left')
# This controls the legend (color explanation box).
# title = 'Network provider'  legend title.
# bbox_to_anchor = (1, 1)  moves legend outside the graph.
# loc = 'upper left'  legend starts from the top-left corner of that position.
# Result: legend doesnt cover the graph.

plt.xlabel('Call duration(in minutes)')
plt.ylabel('Satisfaction score')
plt.title('Call duration VS satisfaction score')
plt.show()
```

## Categorical plot
##### bar plot, box plot, violin plot etc.

```python
# count of customers by network provider
plt.figure(figsize = (10, 5))
sns.countplot(data = df, x = 'Network_Provider', hue = 'Network_Provider', palette = 'Set2', legend = False)  #palette = 'Set2' Chooses a color theme.'Set2' gives soft and pleasant colors
plt.title('Customers count for each provider')
plt.xlabel('Network Provider')
plt.ylabel('Count of Customers')
#plt.legend().remove()
plt.show()
```

## Distribution plots
##### helps understand data distribution, skewness and outlier


```python
plt.figure(figsize = (8, 4))

sns.histplot(df['Call_Duration'], bins = 16, kde = True, color = 'red')
# kde = kernel density estimate

plt.title('Call duartion for each customers')
plt.xlabel('Call Duration(in mins)')
plt.ylabel('Frequency')
plt.show()
```

## Matrix Plot
##### shows relationship between multiple variables
##### used heatmap and clustermap

```python
plt.figure(figsize = (10, 5))
# compute numeric columns only
numeric_data = df.select_dtypes(include = ['number'])
#Creates a new DataFrame containing only numerical columns (like integers or floats) from the original data, as correlation cannot be computed on text.

sns.heatmap(numeric_data.corr(), annot = True, cmap= 'coolwarm', fmt = '.2f', linewidth = 0.5)
# numeric_data.corr(): Calculates the Pearson correlation coefficient between all numeric columns.
# annot = True: Displays the correlation numerical values inside each cell.
# cmap= 'coolwarm': Uses a blue-to-red color map (blue for negative, red for positive correlation).

plt.title('Correlation between Call Duration and Satisfaction Score')
plt.show()
```

```python
# correlation between call du and sat. score is -.75, indicating a strong negative correlation
# this means, if call du incs. then satis. score tends to be decreases 
```

## Box Plot
##### Satisfactory score by netwirk provider

```python
plt.figure(figsize = (8, 5))
sns.boxplot(data = df, x = 'Network_Provider', y = 'Satisfaction_Score', hue = 'Network_Provider', palette = 'Set2', legend = False)
plt.xlabel('Network Provider')
plt.ylabel('Satisfaction score')
plt.title('Satisfaction Score by Provider')
plt.show()
```

## Violin Plot

```python
plt.figure(figsize = (10, 5))
sns.violinplot(data = df, x = 'Network_Provider', y = 'Call_Duration', hue = 'Network_Provider', legend = False)
plt.title('Call Duration VS Network Provider')
plt.xlabel('Network provider')
plt.ylabel('Call Duration')
plt.show()
```
