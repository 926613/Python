### Step2
##### import numpy

```python
import numpy as np
```

```python
# Create numpy array ID with elements from 1 to 20
ID = np.arange(1, 21)
ID
```

###### 1(a). Cal. sum, mean, median and std

```python
print(f'Sum of all elements are: {np.sum(ID)}')
print(f'Mean of array: {np.mean(ID)}')
print(f'Median of array: {np.median(ID)}')
print(f'Standard deviation of array(ID): {np.std(ID)}')
```

###### 1(b). Find the indices of elements greater than 10

```python
idx = np.where(ID > 10) # return index of each element which is greater than 10
print(idx)
```

###### 2(a) Create 2d array of shape 4*4

```python
arr = np.arange(1, 17).reshape(4, 4)
print('2D Array of shape(4*4)\n', arr)
```

###### 2(b) find transpose of array

```python
print('Transpose of array is:\n', np.transpose(arr))
```

###### 2(c) cal. row-wise and column-wise sum 

```python
print(f'sum of elements for each row: {np.sum(arr, axis = 1)}')
print(f'sum of elements for each column: {np.sum(arr, axis = 0)}')
```

###### 3(a) create 2 Array (3*3) with random numbers from 1 to 16. And perform elements-wise addition, sub, and multiplication

```python
arr1 = np.random.randint(1, 16, (3, 3))
arr2 = np.random.randint(1, 16, (3, 3))
print(arr1)
print(arr2)
```

```python
print('Addition:\n', arr1+arr2)
print('Subtraction:\n', arr1-arr2)
print('Multiplication:\n', arr1*arr2)
```

###### 3(b) compute the dot product of the arrays

```python
print(np.dot(arr1 ,arr2))
```
