## Step2: import pandas

```python
import pandas as pd
```

###### 1(a) Create dataframe and print first five rows 

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}
df = pd.DataFrame(data)
df
```

```python
df.head() # prints first five row of dataframe
```

###### 1(b) get summary statistics of age and salary columns

```python
df[['Age', 'Salary']].describe()
```

###### 1(c). Cal. avg salary for HR department

```python
df1 = df[df['Department'] == 'HR']
df1
```

```python
print('Average salary for HR department: ', df1['Salary'].mean())
```

###### 2. Add new column name 'Bonus', which is 10% of the salary

```python
df['Bonus'] = df['Salary']*(10/100)
```

```python
df
```

###### 3. Filter dataframe to show employees aged between 25 and 30

```python
df[(df['Age'] > 25) & (df['Age'] < 30)]
```

###### 4. Group by data by department and cal. the avg. salary for each depatment

```python
avg_salary_foreachdept = df.groupby('Department')['Salary'].mean()
print('Average salary for each department: \n', avg_salary_foreachdept)
```

###### 5. Sort the dataframe by Salary in asc order and save the result to a new csv file.

```python
sort_df = df.sort_values(by = 'Salary', ascending = True)
```

```python
sort_df
```
