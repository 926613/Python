## Step2: import library

```python
import matplotlib.pyplot as plt
```

###### 1(a) create simple line plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y, marker = 'o', linestyle = '-', color = 'r')
```

###### 1(b) Add title, axis labels and a grid

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x, y, marker = 'o', linestyle = '-', color = 'r', label = 'Increasing Trend')
plt.title('Line plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

###### 2(a). Create bar graph by using given data

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color = ['red', 'blue', 'yellow', 'purple'])
plt.xlabel('Student')
plt.ylabel('Marks')
plt.show()
```

###### 2(b). Customize color and add title

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students, marks, color = ['green', 'pink', 'brown', 'grey'])
plt.xlabel('Student')
plt.ylabel('Marks')
plt.title('Students VS marks')
plt.show()
```

###### 3(a). Create pie chart for given data

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10] 
plt.pie(revenue, labels = regions, autopct = '%.1f%%', colors = ['red', 'blue', 'purple', 'yellow'])
plt.title('Total Revenue by continents')
plt.show()
```

###### 3(b). Highlight the region with the highest revenue

```python
import numpy as np
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10] 
# index of highest value
max_index = np.argmax(revenue)
# creating explode list
explode = [0] * len(revenue) # creating list of zero elements of size 4
explode[max_index] = 0.1

plt.pie(revenue, labels = regions, explode = explode, autopct = '%.1f%%', colors = ['red', 'blue', 'purple', 'yellow'])
plt.title('Total Revenue by continents')
plt.show()
```

###### 4. Generate a histogram to show the frequency distribution of randomly generated integers between 1 and 100 (sample size = 1000)

```python
rgn = np.random.default_rng()
ran_num = rgn.integers(1, 100, 1000)
```

```python
plt.hist(ran_num, bins = 10, color = 'blue', edgecolor = 'black')
plt.title('Frequency Distribution')
plt.xlabel('Numbers')
plt.ylabel('Count')
plt.show()
```
