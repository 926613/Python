### Customer feedback analysis in retail industry using numpy

```python
import numpy
```

```python
data = np.array([['C001', 4, 'Excellent service'],
                ['C002', 5, 'Very good customer support'],
                ['C003', 1, 'Very poor service'],
                ['C004', 2, 'Poor customer support'],
                ['C005', 3, 'Average service'],
                ['C006', 4, 'Discount are good'],
                ['C007', 3, 'Avergae service'],
                ['C008', 4, 'Fast and clean service']])
```

```python
print(data)
print(type(data))
```

```python
# total feedback
ratings = data[:, 1].astype(int) # convert rating into integers
print(ratings)
print(f'Total ratings are: {len(ratings)}')
```

```python
# avg rating
print(f'Average ratings: {np.mean(ratings)}')
```

```python
# highest rating
print(f'Highest rating is: {np.max(ratings)}')
```

```python
# minimum ratind
print(f'Minimum rating is: {np.min(ratings)}')
```

```python
unique_rating, rating_counts = np.unique(ratings, return_counts = True)
```

```python
print(f'Unique ratings: {unique_rating}')
print(f'Rating Counts: {rating_counts}')
```

```python
print('Rating Distributions')
for rating, counts in zip(unique_rating, rating_counts):
#The Python zip() function is a built-in utility that takes one or more iterables (like lists, tuples, or strings) and 
#aggregates their corresponding elements into a single iterator of tuples
    print('Rating ',rating, ': ', counts, ' Customers' )
   # print(rating, counts)
```

```python
feedback_cust = data[:, 2]
print('Customers feedback analysis')
print()
for feed in feedback_cust[ : 5]:
    print('-', feed)
```
