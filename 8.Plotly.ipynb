# Ploty : used for visualisation but it creates interactive graphs

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('tourism_data.csv')
df.head()
```

## Line Plot

```python
# Tourist trends over time
fig = px.line(df, x = 'Month', y = 'Tourist_Arrivals', title = 'Monthly Tourist Arrivals')
fig.show()
```

## Bar Plot

```python
fig1 = px.bar(df, x= 'Country', y = 'Tourist_Arrivals', title = 'Tourist Arrivals by Country')
fig1.show()
```

## Scatter PLot

```python
fig2 = px.scatter(df, x = 'Average_Travel_Cost', y = 'Hotel_Bookings', title = 'Bookings VS Avg_Travel_Cost')
fig2.show()
```

## Histogram
##### -shows frequency distribution of numeric data

```python
fig3 = px.histogram(df, x = 'Hotel_Price', title = 'Dist. hotel price')
fig3.show()
```

## Violin Plot

```python
fig4 = px.violin(df, x = 'Season', y = 'Average_Travel_Cost', box = True, title = 'Seasonal Variatios')
fig4.show()
```

## Gannt Plot:
##### -visualise schedules, making it ideal for trip planning

```python
fig5 = px.timeline(df, x_start = 'Start_Date', x_end = 'End_Date', y = 'Activity', title = 'Tourism')
fig5.show()
```

## 3D plot
##### -3D scatter plot provides a multidimensions view of relationship

```python
fig6 = px.scatter_3d(df, x = 'Average_Travel_Cost', y= 'Hotel_Rating', z = 'Hotel_Bookings', 
                     title = '3D analysis of Average cost, Hotel rating and Booking', color = 'Hotel_Rating')
fig6.show()
```
