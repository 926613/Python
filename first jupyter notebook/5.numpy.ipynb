### Array Operations

```python
a = np.array([1, 2, 3])
b = np.array([2, 4, 6])
print(a+b)
print(a**2)
print(a/b)
```

```python
print(a+4)
```

```python
x = np.array([[1, 2, 3], [4, 5, 6]])
y = np.array([[1, 2, 3], [4, 5, 6]])
print(x+y)
print(x+10)
print(x*y)
```

```python
print(x[1, 1])
```

```python
print(x[0, :])
```

```python
print(x[ :, ])
```

```python
print(x[:, -1])
```

```python
print(x)
```

### Reshaping
##### reshape(rows, columns): used to change the dimensions of an array in numpy

```python
reshaped = np.arange(12).reshape(2, 6)
print(reshaped)
```

## Functions of Numpy

```python
arr = np.array([3, 1, 4, 8, 2, 9, 7])
print(f'Mean: {np.mean(arr)}')
```

```python
print(f'Standard Deviations: {round(np.std(arr), 2)}')
```

```python
print(np.median(arr))
```

```python
print(np.sort(arr))
```

```python
# linspace(start, stop, num):
# used to create array in numpy or generates sequence of evenly spaced-values
print(np.linspace(1, 2, 5))
# start: sequence starts value
# stop: sequence stop value and it is include in result
# num: specify samples means interval from 1 to 5 divides, how many parts
```

```python
ran = np.random.randint(1, 4, (3, 3))
```

```python
print(ran)
```

```python
nd = np.arange(1, 26).reshape(5, 5)
print(nd)
```

```python
print(nd[1:4, 1:4])
```

```python
print(nd[:, :])
```

```python
print('Row-wise sum: ', np.sum(nd, axis = 1))
```

```python
print('Column-wise sum: ', np.sum(nd, axis = 0))
```

# CASE STUDY

```python
claims = np.array([[10000, 12000, 15000], # for 1 holders
                  [3000, 9000, 20000],    # 2
                  [90000, 11000, 25000]]) # 3
```

```python
print(f'Claims done by policy holder 3 in month 2: {claims[2, 1]}')
```

```python
print('Total claims per holders')
print(np.sum(claims, axis = 1))
```

```python
# avg claims per month
```

```python
print('Avg claims per month', np.mean(claims, axis = 0))
```

```python
# claims above 15000
high_claims = claims[claims > 15000]
print(high_claims)
```
