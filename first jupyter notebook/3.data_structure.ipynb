```python
feature1 = {'ABS', 'Airbags', 'Sunroof', 'Touchscreen'}
feature2 = {'ABS', 'Airbags', 'Sunroof', 'Cruise Control'}
```

```python
# commom features
com_feat = feature1.intersection(feature2)
print(com_feat)
```

```python
# total features - unique feature of both cars
com = feature1.union(feature2)
print(com)
```

```python
# difference
com_in_car1= feature1.difference(feature2)
print(com)
```

```python
com1 = feature1.symmetric_difference(feature2)
```

```python
print(com1)
```
