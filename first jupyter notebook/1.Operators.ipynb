```python
# item details
item1_price, item2_price = 100, 50
item1_quantity, item2_quantity = 2, 4
```

```python
#cal.
total1 = item1_quantity * item1_price
total2 = item2_quantity * item2_price
```

```python
total_price = total1 + total2
```

```python
# discount if any
discount = 30
```

```python
final_price = total_price - discount

print(f"Total price: {total_price}")
print(f"Discount: {discount}")
print(f"final price: {final_price}")
```

```python
# BITWISE OPEARTOR

# 1. AND

a = 3 # in binary 0011
b = 4 # in binary 0100

# bitwise and compare each bits position if 1 & 1 = 1 or 1 & 0 = 0
#   0 0 1 1
#   0 1 0 0 
#   --------
#   0 0 0 0  -- it is converted into integer = 0
print('Bitwise and: ', a&b)


a = 3 # in binary 0011
b = 4 # in binary 0100
print('Bitwise or:', a|b)

```
