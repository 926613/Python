### Employment Management System

```python
employees = {}
```

```python
def add_employee(emp_id, name, age, department, salary):
    if emp_id in employees:
        print('This emp-id already esixt')
    else:
        employees[emp_id] = {'Name': name, 'Age': age, 'Department': department, 'Salary': salary}
        print('Employee added successfully')
```

```python
def view_employee():
    print('Emp_id\tName\tAge\tDepartment\tSalary')
    for em_id, details in employees.items():
        print(em_id,
              details['Name'],
              details['Age'],
              details['Department'],
              '  ',
              details['Salary'], sep='\t')
```

```python
def search_employee(emp_id):
    if emp_id in employees:
        print('Details of employee')
        print(f'Name: {employees[emp_id]['Name']}')
        print(f'Age: {employees[emp_id]['Age']}')
        print(f'Department: {employees[emp_id]['Department']}')
        print(f'Salary: {employees[emp_id]['Salary']}')
    else:
        print('Employee not found')
```

```python
print('------------Welcome into EMS------------')
flag = True
while (flag):
    print('\n')
    print('1.Add Employee\n2.View all Employees\n3.Search for employee\n4.Exist')
    print('\n')
    user_choice = int(input('Enter your choice(1-4): '))
    if user_choice == 1:
        print('For adding employee enter the following details:')
        emp_id = int(input('Enter employee id: '))
        name = input('Enter name: ')
        department = input('Enter department: ')
        age = int(input('Enter age: '))
        salary = int(input('Enter salary'))
        add_employee(emp_id, name, age, department, salary)
    elif user_choice == 2:
        print('Printing all employees details: ')
        view_employee()
    elif user_choice == 3:
        print('search for a particular employee: ')
        emp_id = int(input('Enter the emp_id: '))
        search_employee(emp_id)
    elif user_choice == 4:
        flag = False
    else:    
        print('Invaild input')
print('Program ends.')        
print("Thanks you")        
```
