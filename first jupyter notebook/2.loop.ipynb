```python
# sum of numbers from 1 to n
n = int(input("Enter number: "))
i = 1
sum = 0
while i <= n:
    sum += i
    i += 1
print(f"Sums of numbers from 1 to {n} is: {sum}")    
```

```python
# factorial
i = 1
fac=  1
num = int(input("Enter number: "))
while i <= num:
    fac *= i
    i += 1
print(f"Factorial of {num} is: {fac}")    
```

```python
for i in range(3):
    
    for j in range((2*i)+1):
        print('*', end='')
    print()    
```

```python
rows= int(input("Enter numbers of rows: "))
for i in range(1, rows+1):
    # print(i)
    for j in range(rows- i):
        print(' ', end= '')
    for k in range(i):
        print('*', end= '')
    print()    
```

```python
rows= int(input("Enter numbers of rows: "))
for i in range(0, rows):
    # print(i)
    for j in range(rows- i):
        print(' ', end= '')
    for k in range((2*i)+1):
        print('*', end= '')
    print()  
```

```python
sum= 0
for i in range(1, 3):
    for j in range(1, 4):
        num = int(input(f"Enter element at [{i}, {j}]: "))
        sum += num
print(f'Sum of total element in 3*3 matrix is: {sum}')        
```

```python
# break
for i in reversed(range(1, 11)):
    if (i == 6):
        break
    print(i)    
```

```python
# continue
for i in range(1, 11):
    if (i == 6):
        continue
    print(i)    
```
