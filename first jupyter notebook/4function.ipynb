### Recursive function

```python
def factorial(n):
    if (n == 1):
        return 1
    else:
        return n * factorial(n-1)
```

```python
num = int(input("Enter number: "))
x = factorial(num)
print(x)
```

```python
def fibo(n):
    if (n <= 0):
        return 0
    elif n== 1:    
        return 1
    else:
        return fibo(n-2) + fibo(n-1)
```

```python
print(fibo(6))
```

```python
def sum_nat(n):
    if n <= 1:
        return 1
    else:
        return n + sum_nat(n-1)
```

```python
print(sum_nat(10))
```

```python
# map() : it is built in function that applies a function to each item in an iterable
```

```python
num = [1, 2, 3, 4, 5, 6, 7]
def double_func(x):
    return x*2

re = map(double_func, num)
print(f"Result: {list(re)}")
```

```python
sq = map(lambda x: x**2, num)
print(list(sq))
```

```python
fru = ['apple', 'grapes']
up = map(lambda x:x.upper(), fru)
print(list(up))
```

### CASE STUDY
#### Bank Customer Transaction Management System using function

```python
customers = {}
```

```python
def add_customer(account_num, name, balance):
    if account_num in customers:
        print('Account already esixt.')
    else:    
        customers[account_num] = {'Name': name, 'Balance': balance}
        print('Customer added successfully')
```

```python
def display_details(account_num):
    if account_num in customers:
        print(f'Account number: {account_num}')
        print(f'Name: {customers[account_num]['Name']}')
        print(f'Balance: {customers[account_num]['Balance']}')
    else:
        print('Not Found.')
```

```python
def update_account_detail(account_num, amount, transaction_type):
    if account_num in customers:
        if transaction_type == 'deposit':
            customers[account_num]['Balance'] += amount
        elif transaction_type == 'withdrawl' and customers[account_num]['Balance'] >= amount: 
            customers[account_num]['Balance'] -= amount
        else:    
            print('Insufficient balance')
            return 0
        print('Transaction successfully')
    else:    
        print('Account not found')
```

```python
print('*****Enter into Banking System******')
flag = True
while (flag):
    print('\n')
    print('1.Add customer\n2.Display detaila\n3.Update\n4.exit')
    print('\n')
    user_choice = int(input("Enter your choice(1-4)"))
    if user_choice == 1:
        account_num = int(input('Enter account number: '))
        name = input('Enter account holder name: ')
        balance = int(input('Enter account balance: '))
        add_customer(account_num, name, balance)
    elif user_choice == 2:
        account_num = int(input('Enter account number: '))
        display_details(account_num)
    elif user_choice == 3:
        print('For updation enter the following details: ')
        account_num = int(input('Enter account number: '))
        amount = int(input('Enter amount(deposit/withdrawl): '))
        transaction_type = input('Enter transaction type: ')
        update_account_detail(account_num, amount, transaction_type)
    elif user_choice == 4:    
        flag = False
    else:
        print('Invalid choice')
print('program end.')        
```
