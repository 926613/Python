```python
pur_amount = int(input("Enter amount: "))
if (pur_amount >= 5000):
    discount = .2 * pur_amount
    print("20% off")
    print(f"Discount: {discount}")
elif (pur_amount >= 3000):
    discount = .1 * pur_amount
    print("10% off")
    print(f"Discount: {discount}")
else:
    discount = 0
    print("NO discount")
Final_price = pur_amount - discount
print(f"Final price: {Final_price}")    
```
