```python
try:
    ## code block where exception can occur
    z = d
except NameError as ex1:
    print("The user don't defined the variable")
    
except Exception as ex:
    print(ex)


# The except Exception as ex block catches any type of exception that inherits from the base Exception class, including NameError.

# print(ex): The code inside the except block executes, 
# printing the specific error message associated with the NameError, which is "name 'd' is not defined".     
```

```python
x = 1
c = 's'
d = x+c
```

```python
try:
    x = 1
    c = 's'
    d = x+c

except TypeError:    
    print('Try to enter same type of datatype')
except Exception as ex:
    print(ex)
```

```python
try:
    a= 12
    b = 0
    c = a/b
except TypeError:    
    print('Try to enter same type of datatype')
except ZeroDivisionError:
    print('Divide by zero is not possible.')
except Exception as ex:
    print(ex)
```

```python
## try else block
try:
    a= int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = a/b
except TypeError:    
    print('Try to enter same type of datatype')
except ZeroDivisionError:
    print('Divide by zero is not possible.')
except Exception as ex:
    print(ex)

else:
    print('Result: ', c)

# The else block runs only when NO exception occurs
 
# * If any exception is raised in the try block, the else block is skipped
# * If no exception occurs, the else block executes    
```

```python
## try else finally

try:
    a= int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = a/b
except TypeError:    
    print('Try to enter same type of datatype')
except ZeroDivisionError:
    print('Divide by zero is not possible.')
except Exception as ex:
    print(ex)

else:
    print('Result: ', c)

finally:
    print('Program end here.')
# try-finally block is used to ensure that specific cleanup code always executes, regardless of whether an exception occurred or not.
    
```

# Custom Exception

```python
class Error(Exception):
    pass
class NegativeNumberError(Error):
    pass
```

```python
num = int(input('Enter number: '))
try:
    if num > 0:
        print('Thanks for feedback.')
    else:
        raise NegativeNumberError('Do not enter negative number for feedback.')
except NegativeNumberError as e:
    print(e)
    print('Feedback is not in negative numnber.')
```
