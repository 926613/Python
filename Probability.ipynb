## Basis probability: tossing a coin

```python
from fractions import Fraction 
# The from fractions import Fraction statement in Python imports the Fraction class from the built-in fractions module, 
# allowing for rational number arithmetic (fractions) without floating-point precision errors.
# it is especially used in prob. cal.
```

```python
# define outcomes and sample spaces
outcomes = ['head', 'tail']
total_outcomes = len(outcomes)

# cal. prob.
p_h = Fraction(outcomes.count('head'), total_outcomes)
p_t = Fraction(outcomes.count('tail'), total_outcomes)

print('Sample: ', outcomes)
print('Sample size: ', total_outcomes)
print('P(Head): ', p_h)
print('P(Tail): ', p_t)
```

## Baye's Theorem: Case Study

```python
# let D = has Diseases
#     T = positive test

P_D = 0.001
P_T_given_D = 0.99
P_T_given_no_D = 0.05
P_no_D = 0.999

numerator = P_D*P_T_given_D
denomenator = (P_D*P_T_given_D)+(P_no_D*P_T_given_no_D)
P_D_given_T = numerator/denomenator

print(f'P_D_given_T: {P_D_given_T:.4f} (~{round(P_D_given_T, 4)}%)')
```

## Tossed 3 coin: Find prob. of getting exactly 2 heads

```python
from itertools import product
# product function  in Python is used to find all possible combinations of given sample
# with leght = 3

# product(['H', 'T'], repeat = 3): This creates an iterator that finds all possible combinations of 'H' and 'T' with a length of 3.
com = list(product(['H', 'T'], repeat = 4))
# list(...): The product function returns an iterator (a generator).
# We use list() to immediately convert that generator into a list of tuples so we can see all results at once.
com
```

```python
fav_out = [i for i in com if i.count('H') == 2]
print('Total outcomes: ', len(com))
print('Toatal fav outcomes: ', len(fav_out))
print('P(getting exactly 2 heads): ', Fraction(len(fav_out), len(com)))
```

## Dice rolls eg. : sum 7

```python
total_out = list(product(range(1, 7), repeat = 2))

fav = [i for i in total_out if sum(i) == 7]

print('Total outcomes: ', len(total_out))
print('Total fav outcomes', len(fav))
print('P(getting sum 7): ', Fraction(len(fav), len(total_out)))
```
