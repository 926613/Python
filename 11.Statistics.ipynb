# Descriptive Statistics

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame(
    {
        'Price': [500000, 900000, 340000, 1300000, 2500000, 19000000],
        'Car_color': ['Red', 'White', 'Red', 'Black', 'Black', 'Grey']
    }
)
data
```

### Mean

```python
mean_price = data['Price'].mean()
print('Mean price of cars:', mean_price)
```

### Median

```python
median_price = data['Price'].median()
median_price
```

### Mode

```python
mode_color = data['Car_color'].mode()
mode_color
```

### Standard Deviation

```python
std_price = data['Price'].std()
std_price
```

### Variance

```python
var_price = data['Price'].var()
var_price
```

### Range

```python
range_price = data['Price'].max() - data['Price'].min()
range_price
```

### Percentile and Quantile

```python
Q1 = data['Price'].quantile(.25)
Q2 = data['Price'].quantile(.50)
Q3 = data['Price'].quantile(.75)
print('Q1', Q1)
print('Q2', Q2)
print('Q3', Q3)
```

```python
### Boxplot
sns.boxplot(y = data['Price'])
plt.show()
```

```python
sns.histplot(data['Price'], kde = True)
plt.show()
```

# EXAMPLE

```python
data1 = pd.DataFrame(
    {
        'INFY': [5.2, 4.8, 5.5, 5.0, 4.9, 5.3],
        'Reliance': [8.5, 12.2, 7.8, 15.3, 10.1, 9.0],
        'HDFCBank': [6.1, 5.5, 6.7, 6.2, 5.8, 6.0]
    }
)
```

```python
data1
```

```python
# mean
mean_val = data1.mean()
```

```python
mean_val
```

```python
# Risk cal.
std_val = data1.std()
std_val
```

```python
data1.plot(kind = 'box', title = 'INFY VS Reliance VS HDFCBank')
plt.ylabel('Monthly Return')
plt.grid()
plt.show()
```

```python
# inference: monthly return of Reliance is higher than other companies
#            mean spread for Relaince is also higher and volatility is high, means risk is also high

# INFY and HDFC : offers stable data or you can say consistent. Risk is very low
```
